```python
!pip install anthropic
import anthropic
from google.colab import userdata

client = anthropic.Anthropic(api_key=userdata.get('Anthropic'))

def ask(prompt):
    msg = client.messages.create(
        model="claude-sonnet-4-5",
        max_tokens=1000,
        messages=[{"role": "user", "content": prompt}]
    )
    return msg.content[0].text

print("Setup done. Claude is connected.")
```

```python
ticket = "I was charged twice for my subscription and no one has replied to my email in 3 days."

prompt = f"""You are a customer support triage assistant.
Classify this ticket. Return exactly these 4 lines:
Category: (billing / technical / complaint / other)
Sentiment: (one word)
Escalate to human: (yes/no)
Reason: (one short sentence)

Ticket: {ticket}"""

print(ask(prompt))
```

```python
def classify_ticket(ticket):
    prompt = f"""You are a customer support triage assistant.
Classify this ticket. Return exactly these 4 lines:
Category: (billing / technical / complaint / other / Spam)
Sentiment: (one word)
Escalate to human: (yes/no)
Reason: (one short sentence)
Suggested reply: (one sentence)

Ticket: {ticket}"""
    return ask(prompt)

# test it
print(classify_ticket("My app keeps crashing every time I open the reports page."))
```

```python
tickets = [
    "I want a refund, I was double charged!",
    "How do I export my data to CSV?",
    "Your service is terrible and I'm cancelling.",
    "asdfghjkl",  # deliberate gibberish to test edge cases
    "I love the product but the mobile app is missing dark mode."
]

for t in tickets:
    print("TICKET:", t)
    print(classify_ticket(t))
    print("-" * 40)
```
